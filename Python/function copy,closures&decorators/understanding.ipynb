```python
## Closures

def main_welcome(func):
    
    def sub_welcome():
        print("Hey Welcome to the Python course")
        func("welcome")
        print("Please learn this concept properly")

    return sub_welcome()

```

```python
main_welcome(print)
```

```python
## Closures

def main_welcome(func,list):
    
    def sub_welcome():
        print("Hey Welcome to the Python course")
        print(func(list))
        print("Please learn this concept properly")

    return sub_welcome()

```

```python
main_welcome(len,[1,2,3,4,5])
```

```python
## Decorators

def main_welcome(func):
    
    def sub_welcome():
        print("Hey Welcome to the Python course")
        func()
        print("Please learn this concept properly")

    return sub_welcome()
```

```python
def course_introduction():
    print("This is a advanced python course")


course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
# without using main_welcome we can print which is by using decorators

@main_welcome     # by writing this the second function becomes the frst function parameter
def course_introduction():
    print("This is a advanced python course")
```

```python
def emphaize_decorator(func):
    def wrapper():
        value=func()
        return value**2
    return wrapper

@emphaize_decorator

def square():
    return 3

print(square())
```

```python
def emphasize_decorator(func):
    def wrapper():
        return func().upper() + "!!!"
    return wrapper

@emphasize_decorator
def greet():
    return "hello"

print(greet())
```

```python
# Decorator ex2
def my_decorator(func):
    def wrapper():
        print("something is happening before the function is called")
        result=func()
        print("something is happening after the function is called")
        return result
    return wrapper
```

```python
@my_decorator

def say_hello():
    return "hello"

print(say_hello())
```

```python
class Driver:
    def __init__(self,name,gender):
        # self.__name=name
        # self.__gender=gender
        self.name = name
        self.gender = gender

    def get_name(self):
        return self.name
    
    def set_name(self,name):
        self.name=name


    def get_gender(self):
        return self.gender

    def set_gender(self,gender):
        self.gender = gender

driver=Driver("Vishu","Male")
print(driver.get_name())
driver.set_name("kiran")
print(driver.get_name())
print(driver.get_gender())
driver.set_gender("Female")
print(driver.get_gender())
```
