```python
class car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} car")
```

```python
car1 = car(4,4,'diesel')
car1.drive()
```

```python
### Multiple Inheritance

class Animal:

    def __init__(self,name):
        self.name = name

    def speak(self):
        print(f"{self.name} speaks harshly")


class pet:
    def __init__(self,owner):
        self.owner = owner


class Dog(Animal,pet):

    def __init__(self, name,owner):
        Animal.__init__(self,name)
        pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says woof!!!!"


w = Dog("alex","Vishweshwar")

print(w.speak())
print(w.owner)
        
```

```python
# PolyMorphism (Same Method With Different Implementations)

class Animal1:

    def speak(self):
        return f"It makes the sound of a animal"
    

class Dog(Animal1):

    def speak(self):
        return "WOOF!!!!!!!"
    

class Cat(Animal1):
    
    def speak(self):
        return "MEOW!!!!"
    
# function that Demonstrates polymorphism

def animal_speak(Animal1):
    print(Animal1.speak())
    

d = Dog()
C = Cat()

print(d.speak())
print(C.speak())

# same output with different implementations
animal_speak(d)
animal_speak(C)
```

```python
## Polymorphism with Functions & Methods
## base class

class Shape:
    def area(self):
        return "The Area of the Figure"
    
# Derived Class 1

class Rectangle(Shape):

    def __init__(self,width,length):
        self.width = width
        self.length = length


    def area(self):
        return self.width * self.length
    

# derived class 2 

class circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14 *self.radius*self.radius
    

## Function that Demostrates Polymorphism

def print_area(Shape):
    print(f"The area is {Shape.area()}")


r = Rectangle(4,5)
cr = circle(3)

print(r.area())
print(cr.area())
# print_area(r)
# print_area(cr)
```

```python
# ENcapsulation with getter and setter

class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age


    def get_name(self):
        return self.__name
    

    def set_name(self,name):
        self.__name=name


    def get_age(self):
        return self.__age
    

    def set_age(self,age):
        if age > 0:
            self.__age = age

        else:
            print("invalid Age")
        

person = Person("krish",90)


# Acess and modify private variables using getter ans setter

print(person.get_name())
print(person.get_age())


person.set_name("Vishu")
print(person.get_name())


person.set_age(65)
print(person.get_age())

person.set_age(0)
print(person.get_age())

```
