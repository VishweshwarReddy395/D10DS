```python
from abc import ABC , abstractmethod
```

```python
class Vehicle(ABC):
    def drive(self):
        print("The Vehicle is used for Driving")


    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("car Engine Started")


def operated_vehicle(Vehicle):
    Vehicle.start_engine()
    Vehicle.drive()


car = Car()
operated_vehicle(car)
```
