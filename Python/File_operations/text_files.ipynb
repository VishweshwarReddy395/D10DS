```python
## read a file line by line

with open('example.txt','r') as file:
    for line in file:
        # print(line)   # read and prints each line and also a newline character
        print(line.strip()) # Removes the newline character
```

```python
# Writing a File(overwriting)

with open('example.txt','w') as file:
    file.write("I scored 55 marks in maths exam\n")
    file.write("I scored 95 marks in social exam\n")

  # the previous data in the file is overwritten by using write
```

```python
# write data into the file without overwriting the previous data 

with open('example.txt','a') as file:
    file.write("Append Operation is taking place!\n")
    file.write("This Operation is very useful because our data is not replaced by writing new data!!\n")
    file.write("Soo Yahooo I liked it!!\n")
```

```python
def count_text(file_path):
    with open(file_path,'r') as file:

        lines = file.read()

        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)

        return line,word_count,char_count
    
file_path = 'example3.txt'

lines , words, characters = count_text(file_path)

print(f'lines : {lines},words = {words} , characters : {characters}')
```
