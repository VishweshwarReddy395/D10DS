```python
def square(n):
    for i in range(3):
        yield i**2

square(3)
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
gen
```

```python
# next(gen)
next(gen)
```

```python
for val in gen:
    print(val)
```

```python
def large_files(filepath):
    with open(filepath,'r') as file:
        # contents = file.read()
        # print(contents)
        for line in file:
            yield line

filepath = "gen_example.txt"
for line in large_files(filepath):
    print(line.strip())
```
