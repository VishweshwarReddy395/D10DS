```python
# instance variable and methods

class Dog:
    # constructor
    def __init__(self,breed,color):
        self.breed = breed
        self.color=color
    
# objects
dog1=Dog("Lab","black")
print(f"The Breed of the Dog is {dog1.breed}")
print(f"The color of the Dog is {dog1.color}")

```

```python
dog2 = Dog("Golden Retriever","Brown")
print(f"The Breed of the Dog is {dog2.breed}")
print(f"The color of the Dog is {dog2.color}")
```

```python
# Define a class with instance methods
class cat:
    def __init__(self,name,color):
        self.name = name
        self.color=color

    def bark(self):
        print(f"{self.name} says meow meow!!!")


c = cat("Luciya","Brown")
c.bark()
```

```python
# modeling a Bank account

class Bank:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"{amount} is deposited.Now Balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print(f"Insufficient Balance")

        else:
            self.balance -= amount
            print(f"{amount} is Withdrawn.New Balance is {self.balance}")

    def get_balance(self):
        return self.balance
    

ac=Bank("krish",5000)
print(ac.balance)
```

```python
ac.deposit(100)
```

```python
ac.withdraw(500)
```

```python
ac.get_balance()
```
