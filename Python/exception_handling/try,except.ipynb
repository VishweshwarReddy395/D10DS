## Exception
#### Exception means errors that occur during program execution which disrupts the flow  of execution


```python
# a = b # throws a name error
```

```python
# so we use try,except blocks

try:
    a = b       # where exception occurs

except:
    print("The Variable is not assigned")
```

```python
try:
    a = b       # where exception occurs

except NameError as ex:
    print(ex)
```

```python
# a = 1/0   ZeroDivisionError: division by zero error occurs    
```

```python
try:
    result = 1/0

except ZeroDivisionError as zde:
    print(zde)
    print('Enter the Denominator greater than zero')
```

```python
# here we are only handling zeroDivision errors not name errors 
# check output for clarification

# try:
#     result = 1/2
#     a = b
# except ZeroDivisionError as zde:
#     print(zde)
#     print('Enter the Denominator greater than zero')
```

```python
# for the above problem , we have used Exception to address multiple exceptions
# but Exception block should execute only after all other exceptions.It is like a parent class from which other exceptions are solved

try:
    result = 1/2
    a = b

except ZeroDivisionError as zde:
    print(zde)
    print('Enter the Denominator greater than zero')

except Exception as ex:
    print(ex)
```

```python
try:
    a = b          # This will raise NameError
    result = 1/0   # This will raise ZeroDivisionError (if reached)

except NameError as e:
    print(f"NameError caught: {e}")

except Exception as e:
    print(f"ZeroDivisionError caught: {e}")

```
