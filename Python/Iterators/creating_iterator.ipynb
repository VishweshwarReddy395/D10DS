```python
## In order to iterate through all the elements
## we use next to iterate the element

# next(iterator)
```

```python
my_list = [1,2,3,4,5,6]
iterator = iter(my_list)
try:
    while True:
        print(next(iterator))

except StopIteration:
    print("There are no elements")
```
