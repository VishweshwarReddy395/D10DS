```python
## creating a Directory

import os

new_directory = "package"    # package named directory is created
os.mkdir(new_directory)

print(f"Directory '{new_directory}' create")
```

```python
# Listing Files and Directories

items = os.listdir('.')     # Dot('.') specifies files and directories in a folder
print(items)
```

```python
# Joining Paths
# we can only get the relative paths using directory name and file name

dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(dir_name,file_name)

print(full_path)
```

```python
# we can get our current working directory 
# we can get the entire path  

dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(os.getcwd(),dir_name,file_name)

print(full_path)
```

```python
# path = "example4.txt"   iis a file
path = "package"     # is a directory

if os.path.isfile(path):
    print(f"The Path '{path}' is a file")

elif os.path.isdir(path):
    print(f"The Path '{path}' is a directory")

else:
    print(f"The Path '{path}' is neither a file nor a directory")
```
