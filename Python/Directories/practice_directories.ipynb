```python
# b = "example.txt"
b = "Modules" 
if os.path.isfile(b):
    print(f"The path '{b}' is a file")
elif os.path.isdir(b):
    print(f"The path '{b}' is a Directory")
else:
    print(f"The path '{b}' is neither a file nor a Directory")
```
